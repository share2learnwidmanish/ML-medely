### Numpy is a multidimensional array library. 
- Numpy is much faster than list, numpy uses fixed type<br>
Eg: In numpy a decimal 5 is stored using 32 bits(int32) while in list to store the same integer it will need Object value(64bits), Object byte(64bits), Reference count(64bits), Size(32 bits)
- Numpy uses contiguous block of memory while for list it may be scattered aorund computers memory
