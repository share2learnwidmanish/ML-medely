### Groupby (with Multilevel index) , cross-sectin
- For categorical column, create a "lazy" groupby object, this lazy objects is evaauated by when aggregate method is called


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Data/mpg.csv")
df.head(2)
```

```python
### Lets start with 'model_year' column, groupby model_year column 
df.groupby(by='model_year') ### Thsi generates the lazy group by objects which needs an aggregate function 
```

```python
df.groupby(by='model_year').mean()
#grouped by model_year with mean for other attributes.
```

```python
### Lets group by multiple colums, multilevel heirarchical index
df.groupby(by=['model_year','cylinders']).mean().head(4)
```

```python
df.groupby(by=['model_year','cylinders']).mean().index
```

```python
df.groupby(by=['model_year','cylinders']).mean().columns
### model_year and cylinders are not columns anymore, there are part of multilevel index
```

```python
year_with_cyl = df.groupby(by=['model_year','cylinders']).mean()
```

```python
year_with_cyl.head(4)
```

```python
year_with_cyl.index.names
```

```python
year_with_cyl.index.levels
```

```python
# These are all the index which can be used to query a subset of data. 
#Note now you have to pass a tuple ( outer index, inner index)
year_with_cyl.index
```

```python
# query with outside index
year_with_cyl.loc[[70,82]]
```

```python
# to get a specific row in a multiindex dataframe
year_with_cyl.loc[[(70,8),(82,6)]]
```

### Lets understand what is cross-section

```python
year_with_cyl.head(4)
```

```python
year_with_cyl.xs(key=70,level='model_year')
```

### We want all the instance of 4 cyclinders for every year

```python
year_with_cyl.xs(key=4, level='cylinders')
```

```python
year_with_cyl.sort_index(level='model_year', ascending=False).head(20)
```
