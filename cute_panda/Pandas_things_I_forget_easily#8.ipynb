### Lets see the assign() method

```python
df.head(2)
```

### Assign with only one transformation 

```python
df = df.assign(tip_per = lambda x : round(x['tip']/x['total_bill'],2))
df.sort_values(by='tip_per')
```

### Assign with multiple transformation 

```python
df = df.assign(tip_perc = lambda x : round(x['tip']/x['total_bill'],2), tip_P_person = lambda x : round(x['tip']/x['size'],2))
df.sort_values(by='tip_per')
```

### Lets look at map function
- Lets transform the day colum into numbers, Sun-0, Mon-0....Sat-6

```python
df['day'].unique()
```

```python
weekdays = {'Sun':0, 'Mon':1, 'Tues':2,'Wed':3,'Thur':4, 'Fri':5,'Sat':6}
```

```python
df['day_num']= df['day'].map(weekdays)
df[df['day'].isin(['Fri'])].head(2)
```

### Lets look at "map" function with lambda function
- Lets mask the credit card number and display only the last four digits

```python
df['CC Number_mask'] = df['CC Number'].map(lambda x : "XXXX"+str(x)[-4:] if len(str(x))==16 else "xxxx")
df.head(4)
```
