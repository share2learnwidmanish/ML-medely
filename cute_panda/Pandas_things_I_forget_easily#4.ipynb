### Lets Rename "CC Number" and "Payment ID" to "CC_Number" and "Payment_ID" respectively

```python
df.head(2)
```

```python
df.rename(columns={"CC Number": "CC_Number", "Payment ID": "Payment_ID"})
```

### Lets format the headers to Title case

```python
df.rename(str.title, axis="columns")
```

### Lets Rename with lambda function

```python
df.columns
```

```python
df.rename(lambda x : str.upper(x), axis=1).head(1)
```

```python
df.rename(lambda x : str.title(x), axis=1).head(1)
```
